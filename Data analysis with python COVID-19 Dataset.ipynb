```python
#showing  the number of confirmed, death and recovered cases in each region use group by function.
covid_19.head(3)
```

```python
#Each, use group by function
covid_19.groupby('Country/Region').sum()
```

```python
covid_19.groupby('Country/Region').sum().head(25)
```

```python
#only confirmed cases.(Alphabetically)columns names are cases sensitive.
covid_19.groupby('Country/Region')['Confirmed'].sum()
```

```python
covid_19.groupby('Country/Region')['Deaths'].sum()
```

```python
covid_19.groupby('Country/Region')['Recovered'].sum()
```

```python
#Maximum number of cases where recorded in region
covid_19.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False)
```

```python
#Minimum number of death where recorded in Region
covid_19.groupby('Country/Region').Deaths.sum().sort_values(ascending=True).head(50)
```
